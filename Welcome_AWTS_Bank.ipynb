```python
print("############################## Welcome_AWTS_Bank #############################")
users = {}    
accounts = {}  
admin = {"admin": "admin123"}  

def admin_login():
    u = input("Admin Username: ")  
    p = input("Admin Password: ")  
    return admin.get(u) == p

def create_account():
    acc_no = input("Enter Account Number: ")  
    if acc_no in accounts:
        print("Account already exists")
        return

    username = input("Enter Username (must be registered): ")  
    if username not in users:
        print("User not found")
        return

    name = input("Account Holder Name: ")  
    balance = float(input("Initial Balance: "))  
    accounts[acc_no] = {
        "username": username,
        "name": name,
        "balance": balance
    }
    print("✅ Account created successfully")

def view_accounts():
    if not accounts:                  
        print("No accounts yet")
        return
    print("\n--- Bank Accounts ---")
    for acc, info in accounts.items():   
        print(f"Acc: {acc} | User: {info['username']} | Name: {info['name']} | Balance: {info['balance']}")

def signup():
    username = input("Choose Username: ")  
    if username in users:                  
        print("Username already exists")
        return
    password = input("Choose Password: ") 
    users[username] = password            
    print("✅ Signup successful")

def signin():
    u = input("Username: ")              
    p = input("Password: ")              
    if users.get(u) == p:                
        return u                         
    print("Invalid login")
    return None                           

def get_user_account(username):
    for acc, info in accounts.items():   
        if info["username"] == username: 
            return acc, info             
    return None, None                     

def deposit(username):
    acc, info = get_user_account(username)  
    if not acc:                             
        print("No account assigned")
        return
    amount = float(input("Deposit Amount: "))  
    info["balance"] += amount                   
    print("✅ Deposit successful")

def withdraw(username):
    acc, info = get_user_account(username)     
    if not acc:                                
        print("No account found")
        return
    amount = float(input("Withdraw Amount: ")) 
    if amount > info["balance"]:               
        print("❌ Insufficient balance")
    else:
        info["balance"] -= amount             
        print("✅ Withdrawal successful")

def transfer(username):
    send_acc, sender = get_user_account(username)  
    if not send_acc:
        print("No account found")
        return
    rec_acc = input("Receiver Account Number: ")   
    if rec_acc not in accounts:                    
        print("Receiver account not found")
        return
    amount = float(input("Amount to transfer: "))  
    if amount > sender["balance"]:                 
        print("❌ Not enough balance")
        return
    sender["balance"] -= amount
    accounts[rec_acc]["balance"] += amount
    print("✅ Transfer successful")

def view_account(username):
    acc, info = get_user_account(username)  
    if not acc:
        print("No account found")
        return
    print("\n--- Account Details ---")
    print("Account No:", acc)
    print("Name:", info["name"])
    print("Balance:", info["balance"])

while True:  
    print("\n======= BANK SYSTEM =======")
    print("1. Admin Login")
    print("2. User Signup")
    print("3. User Login")
    print("4. Exit")

    choice = input("Select option: ")

    if choice == "1":  
        if admin_login():  
            while True:
                print("\n-- Admin Menu --")
                print("1. Create Account")
                print("2. View Accounts")
                print("3. Logout")

                ch = input("Choose: ")
                if ch == "1":
                    create_account()  
                elif ch == "2":
                    view_accounts()   
                elif ch == "3":
                    break             
        else:
            print("Admin login failed")

    elif choice == "2":  
        signup()

    elif choice == "3":  
        user = signin()
        if user:
            while True:
                print("\n-- User Menu --")
                print("1. Deposit")
                print("2. Withdraw")
                print("3. Transfer")
                print("4. View Account")
                print("5. Logout")

                ch = input("Choose: ")
                if ch == "1":
                    deposit(user)
                elif ch == "2":
                    withdraw(user)
                elif ch == "3":
                    transfer(user)
                elif ch == "4":
                    view_account(user)
                elif ch == "5":
                    break

    elif choice == "4":  
        print("Thanks for using bank")
        break

    else:
        print("Invalid option")
```
